```python
# 2. Target Variable: Attrition
# ------------------------------
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Attrition", palette="Set2")
plt.title("Attrition Distribution")
plt.show()

attrition_rate = df["Attrition"].value_counts(normalize=True) * 100
print("Attrition Rate (%):\n", attrition_rate)
```

```python
# 3. Numerical Features
# ------------------------------
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Histograms for numerical features
df[numerical_cols].hist(bins=20, figsize=(20,20))
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
#  Correlations

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Attrition by Work-Life Balance 

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="WorkLifeBalance", hue="Attrition", palette="Set2")
plt.title("Attrition by Work-Life Balance")
plt.show()
```

```python
# Attrition by Job Satisfaction

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="JobSatisfaction", hue="Attrition", palette="Set2")
plt.title("Attrition by Job Satisfaction")
plt.show()
```
