```python
sns.boxplot(data=df, x='Power', y='Units_Sold')
plt.title("Units Sold Distribution by Power Type")
plt.show()
```

# Observation:Units Sold Distribution by Power Type

- This chart highlights how sales vary across different power types.
- A narrower box means more consistent sales; a wider box means higher variability.
- If one power type shows higher medians, it indicates stronger sales performance compared to others.
- Outliers reveal months/countries with exceptionally high or low sales for specific power types.
- Comparing box sizes helps identify whether certain power types are stable or unpredictable in demand.
- Overall, this visualization is useful to spot which fuel types dominate the market and where volatility exists.

```python
# Cost vs Units Sold by Model
model_sales = df.groupby('Model')[['Units_Sold','Cost_USD']].mean().reset_index()
sns.scatterplot(data=model_sales, x='Cost_USD', y='Units_Sold', hue='Model')
plt.title("Units Sold vs Average Cost by Model")
plt.show()
```

## Observation : Cost vs Units Sold by Model
- Each dot shows a car model with its average cost and units sold.
- Cheaper models generally achieve higher sales.
- Premium models sit at the top-right but with fewer units sold.
- Some mid-range models strike a good balance of cost and volume.
- Indicates pricing elasticity of BMW models.
- Helps find sweet spots for profitability and demand.

```python
# 5. Monthly Trend of Units Sold
monthly_sales = df.groupby(['Year','Month'])['Units_Sold'].sum().reset_index()
monthly_sales['Month_Year'] = monthly_sales['Month'] + '-' + monthly_sales['Year'].astype(str)

plt.figure(figsize=(12,5))
sns.lineplot(data=monthly_sales, x='Month_Year', y='Units_Sold', marker='o')
plt.xticks(rotation=90)
plt.title("Monthly Trend of Units Sold")
plt.show()
```

### Observations: Monthly Trend of Units Sold
- The line chart highlights seasonality in BMW sales across months and years.  
- Peaks around festive or year-end months (Nov–Dec) show promotional or holiday-driven demand.  
- Certain months like April and August may dip, suggesting off-seasons or external market factors.  
- Year-on-year, the overall sales trend appears upward, confirming market expansion.  
- This pattern is valuable for planning production cycles and inventory allocation.  
- Marketing campaigns could be strategically placed during high-demand months to maximize sales.  


```python
# 7. Revenue vs Profit by Model
sns.scatterplot(data=df, x='Revenue_USD', y='Profit_USD', hue='Model', size='Units_Sold', sizes=(50,300))
plt.title("Revenue vs Profit by Model")
plt.show()
```

### Observations: Revenue vs Profit by Model
- Each point represents a car model with its revenue and profit performance.  
- Some models generate very high revenue but only modest profits, showing thin margins.  
- Others balance moderate revenue with strong profit ratios, making them highly efficient.  
- Bubble sizes reveal which models contribute the most units sold to the business.  
- Premium models typically sit on the right with higher revenue but fewer units sold.  
- This analysis helps identify which models are most profitable versus those simply driving volume.  


```python
# 8. Pairplot of Key Metrics
sns.pairplot(df[['Units_Sold','Revenue_USD','Cost_USD','Profit_USD','Profit_Margin','Gross_Margin']])
plt.show()
```

### Observations: Pairplot of Key Metrics
- This matrix shows relationships between units sold, revenue, costs, profits, and margins.  
- Strong correlation is visible between Units Sold and Revenue, which is expected.  
- Profit varies more widely, indicating different cost structures across models.  
- Margins are not consistently aligned with revenue, suggesting pricing inefficiencies.  
- Outliers appear in a few metrics, highlighting unusual cases worth investigating.  
- Overall, it provides a holistic overview of BMW’s operational efficiency.  


```python
# ROI Distribution Across Models
sns.histplot(df['ROI'], kde=True, bins=20)
plt.title("ROI Distribution Across Models")
plt.show()
```

```python
from plotly.subplots import make_subplots
import plotly.express as px

# Prepare data
region_revenue = df.groupby("Region")["Revenue_USD"].sum().reset_index()

# Create subplots with correct specs (domain for pie)
dashboard = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Revenue by Year", "Top Models", "Profit Margin by Region", "Revenue by Region"),
    specs=[[{"type": "xy"}, {"type": "xy"}],
           [{"type": "xy"}, {"type": "domain"}]]  # 👈 pie needs "domain"
)

# Revenue by Year
fig1 = px.bar(df.groupby("Year")["Revenue_USD"].sum().reset_index(),
              x="Year", y="Revenue_USD")
dashboard.add_trace(fig1.data[0], row=1, col=1)

# Top Models
fig2 = px.bar(df.groupby("Model")["Units_Sold"].sum().reset_index().sort_values("Units_Sold", ascending=False).head(10),
              x="Model", y="Units_Sold")
dashboard.add_trace(fig2.data[0], row=1, col=2)

# Profit Margin by Region
fig3 = px.box(df, x="Region", y="Profit_Margin")
dashboard.add_trace(fig3.data[0], row=2, col=1)

# Revenue Share by Region (Pie)
fig4 = px.pie(region_revenue, names="Region", values="Revenue_USD")
dashboard.add_trace(fig4.data[0], row=2, col=2)

# Layout
dashboard.update_layout(height=900, width=1200,
                        title_text="BMW Sales Analysis Dashboard (EDA)")
dashboard.show()
```
