```python
import pandas as pd
import numpy as np
import random

df = pd.read_csv("C:/Users/kusha/OneDrive/Desktop/Capstone Project/Capstone Project/BMW_Sales_2020_2025.csv")

# Numeric columns: replace NaN with a random number within column range
num_cols = ["Units_Sold", "Avg_Price_USD", "Revenue_USD", "Cost_USD", "Profit_USD"]
for col in num_cols:
    df.replace({col: {np.nan: random.randint(int(df[col].min()), int(df[col].max()))}}, inplace=True)

# Categorical columns: replace NaN with a random existing value
cat_cols = ["Power", "Region", "Country"]
for col in cat_cols:
    df.replace({col: {np.nan: random.choice(df[col].dropna())}}, inplace=True)

df.to_csv("BMW_Sales_Filled_Replace.csv", index=False)
```
